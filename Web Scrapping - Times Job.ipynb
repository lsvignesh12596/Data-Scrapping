```python
def get_jobs(link, pages):
  result_df = pd.DataFrame()
  for i in range(1,pages+1):
    site = link.replace("{i}",str(i))
    html_content = requests.get(site).text
    soup = BeautifulSoup(html_content,"lxml")
    jobs = soup.find_all("li", class_="clearfix job-bx wht-shd-bx")
    companyName = []
    Skills = []
    ExpReq = []
    Loc = []
    Pub_on = []
    More_Info = []

    matches = ["few","7","6","5","4","3","2","1","today", "work"]
    for job in jobs:
      published_on = job.find("span",class_="sim-posted").span.text
      # print(published_on)
      if any(match.lower() in published_on for match in matches):
        company_name = job.find("h3", class_="joblist-comp-name").text.replace("  ","").replace("\n","").replace("\r","")
        skills = job.find("span",class_="srp-skills").text.replace("  ","").replace("\n","").replace("\r","")
        experience_required = re.findall("\d.*\d",(job.find("ul")).find("li").text)[0]
        location = job.find("span").text  #job.find_all("span")[1].text #
        more_info = job.find("h2").a['href']

        companyName.append(company_name)
        Skills.append(skills)
        ExpReq.append(experience_required)
        Loc.append(location)
        Pub_on.append(published_on)
        More_Info.append(more_info)
    
    df = pd.DataFrame()
    df['companyName'] = companyName
    df['Skills'] = Skills
    df['ExpReq'] = ExpReq
    df['Loc'] = Loc
    df['Published On'] = Pub_on
    df['More Info'] = More_Info

    result_df = pd.concat([result_df,df])
    result_df.reset_index(drop=True, inplace=True)
  
  return result_df
```

```python
link = "https://www.timesjobs.com/candidate/job-search.html?from=submit&actualTxtKeywords=python&searchBy=0&rdoOperator=OR&searchType=personalizedSearch&luceneResultSize=25&postWeek=60&txtKeywords=python&pDate=I&sequence=1&startPage={i}"
result = get_jobs(link,10)
```
