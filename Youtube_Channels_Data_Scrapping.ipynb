```python
import os
import google.auth
import google.auth.transport.requests
import requests
from google.oauth2.credentials import Credentials
from apiclient.discovery import build
import pandas as pd
import numpy as np
```

```python
# Set API key and search query
API_KEY = ""  #input the API key
QUERY = ["data science","ai", "machine learning", "nlp", "cv", "computer vision"]
channel_ids = []

# Build the YouTube Data API v3 client
youtube = build('youtube', 'v3', developerKey=API_KEY)
```

```python
for query in QUERY:
    # Define the API request to search for channels
    search_request = youtube.search().list(
        part="id",
        q=query,
        type="channel",
        maxResults=100  # Set the maximum number of search results to retrieve
    )

    # Execute the API request and retrieve the response
    search_response = search_request.execute()

    # Extract the channel IDs from the first page of search results
    channel_ids += [result["id"]["channelId"] for result in search_response["items"]]

    while "nextPageToken" in search_response:
        # Define the API request to retrieve the next page of search results
        search_request = youtube.search().list(
            part="id",
            q=query,
            type="channel",
            maxResults=50,
            pageToken=search_response["nextPageToken"]  # Set the page token for the next page of results
        )
        
        # Execute the API request and retrieve the response
        search_response = search_request.execute()
        
        # Extract the channel IDs from the current page of search results
        channel_ids += [result["id"]["channelId"] for result in search_response["items"]]

# # Print the channel IDs
# print(channel_ids)
```

```python
yt_data = pd.DataFrame()
```

```python

# Build the YouTube Data API v3 client
youtube = build('youtube', 'v3', developerKey=API_KEY)

# Loop over the channel IDs and retrieve the channel data
for channel_id in channel_ids:
    channel_request = youtube.channels().list(
        part="snippet,contentDetails,statistics",
        id=channel_id
    )
    channel_response = channel_request.execute()
    # print(channel_response)
    try:
      country = (channel_response['items'][0]['snippet']['country'])
    except:
      country = ""
    yt_data = yt_data.append([[channel_response['items'][0]['snippet']['title'], channel_response['items'][0]['snippet']['description'],
                              country, channel_response['items'][0]['statistics']['subscriberCount'],
                              channel_response['items'][0]['statistics']['videoCount']]])
```

```python
yt_data.columns = ['channel', 'description', 'country','subscriberCount','videoCount']
```

```python
yt_data.reset_index(inplace=True, drop =True)
yt_data
```
